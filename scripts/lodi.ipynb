```python
import pandas as pd
from glob import glob
import numpy as np
import os
```

```python
traj_files = glob('/Users/KevinBu/Desktop/clemente_lab/iclust/data/lodi/*.tsv')
```

```python
for f in traj_files:
    taxa = os.path.splitext(os.path.basename(f))[0]
    whole_df = pd.read_csv(f, sep='\t')
    whole_df = whole_df[['X','Y','ID']]
    os.mkdir('/Users/KevinBu/Desktop/clemente_lab/iclust/data/lodi_trajs/' + taxa + '/')
    taxa_df = pd.DataFrame(columns=['x','y','ID'])
    for index, row in whole_df.iterrows():
        ID = str(row['ID'])
        df = pd.DataFrame(columns=['x','y','ID'])
        df['x'] = np.array([float(x) for x in row['X'].split(',')])
        df['y'] = np.array([float(x) for x in row['Y'].split(',')])
        df['ID'] = ['s' + str(row['ID'])] * len(df)
        taxa_df = taxa_df.append(df)
    taxa_df.to_csv('/Users/KevinBu/Desktop/clemente_lab/iclust/data/lodi_trajs/' + taxa + '/lodi_noninterp_' + taxa + '.txt',sep='\t', index=False)
```
